```python
from sklearn.impute import SimpleImputer
imp_mode=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df1=pd.DataFrame(imp_mode.fit_transform(df))
df1.columns=df.columns
df1
```

```python
df1['rbc']=df1['rbc'].replace(['normal','abnormal'],[1,0])
df1['pc']=df1['pc'].replace(['normal','abnormal'],[1,0])
df1['pcc']=df1['pcc'].replace(['present','notpresent'],[1,0])
df1['ba']=df1['ba'].replace(['present','notpresent'],[1,0])
df1['htn']=df1['htn'].replace(['yes','no'],[1,0])
df1['dm']=df1['dm'].replace(['yes','no'],[1,0])
df1['cad']=df1['cad'].replace(['yes','no'],[1,0])
df1['appet']=df1['appet'].replace(['good','poor'],[1,0])
df1['pe']=df1['pe'].replace(['yes','no'],[1,0])
df1['ane']=df1['ane'].replace(['yes','no'],[1,0])
```

```python
df1['classification']=df1['classification'].replace(['ckd','notckd'],[1,0])
```

```python
df1['age'] = pd.to_numeric(df1['age'])
df1['bp'] = pd.to_numeric(df1['bp'])
df1['sg'] = pd.to_numeric(df1['sg'])
df1['al'] = pd.to_numeric(df1['al'])
df1['su'] = pd.to_numeric(df1['su'])
df1['bgr'] = pd.to_numeric(df1['bgr'])
df1['bu'] = pd.to_numeric(df1['bu'])
df1['sc'] = pd.to_numeric(df1['sc'])
df1['sod'] = pd.to_numeric(df1['sod'])
df1['pot'] = pd.to_numeric(df1['pot'])
df1['hemo'] = pd.to_numeric(df1['hemo'])
df1['pcv'] = pd.to_numeric(df1['pcv'])


```

```python
from sklearn.model_selection import train_test_split

x=df1[['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr',
       'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv','appet', 'pe', 'ane']]
y=df1['classification']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0,max_iter=500).fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report( y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
rf = RandomForestRegressor(n_estimators = 51, random_state = 1)
model = rf.fit(x_train, y_train)
np.sqrt(mean_squared_error(y_train, model.predict(x_train)))
r2_score(y_train, model.predict(x_train))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=23)
dt_model.fit(x_train, y_train)
y_pred_dt = dt_model.predict(x_test)
print(classification_report(y_test, y_pred_dt))
```
